```python
import sys, os
everestPath = os.path.abspath('everest')
if not everestPath in sys.path:
    sys.path.insert(0, everestPath)
```

```python
import pickle
```

```python
with open('nudat.pkl', mode = 'rb') as file:
    nus = pickle.load(file)
```

```python
with open('chrondat.pkl', mode = 'rb') as file:
    chrons = pickle.load(file)
```

```python
from everest.window import Canvas
%matplotlib inline
```

```python
canvas = Canvas(size = (24, 18))
ax = canvas.make_ax()
for nu, chron in zip(nus.values(), chrons.values()):
    if len(nu):
        ax.line(chron, nu)
canvas.show()
```
